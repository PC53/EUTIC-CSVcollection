```python
import yfinance as yf
import pandas as pd
```

```python
# nasdaq_data = pd.read_csv('nasdaq_screener_1668715796429.csv')
# tickers = nasdaq_data['Symbol']
tickers = []
print(f'we have {len(tickers)} tickers')
```

```python
main_dict = {}

for ticker in tickers:
    ticker_df = yf.download(ticker, start="2012-01-01", interval='1d')
    main_dict[f'{ticker}'] = ticker_df['Adj Close']
    
final_df = pd.DataFrame(main_dict)
final_df.to_csv('closing_prices.csv')
```
