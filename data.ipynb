```python
import yfinance as yf
import pandas as pd
import sqlite3 as sq3
```

```python
holdings_data = pd.read_excel('holdings-CONFIDENTIAL.xlsx')
holdings_data.dropna(axis=1)
tickers = holdings_data['yfinance']
tickers
```

```python
main_dict = {}

for ticker in tickers:
    ticker_df = yf.download(ticker, start="2012-01-01", interval='1d')
    main_dict[f'{ticker}'] = ticker_df['Adj Close']
    
final_df = pd.DataFrame(main_dict)
final_df.to_csv('closing_prices.csv')
```
